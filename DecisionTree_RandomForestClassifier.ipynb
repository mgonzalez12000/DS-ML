```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
# Part D imports
from sklearn.utils import resample
```

```python
# A. Read the dataset file “Cancer.csv” (from github using the following command), and assign it to a Pandas 
# DataFrame:
df  = pd.read_csv("https://github.com/mpourhoma/CS4661/raw/master/Cancer.csv")
df
```

```python
# B. Use sklearn functions to split the dataset into testing and training sets with the following 
# parameters: test_size=0.35, random_state=3.
feature_cols = ['Clump_Thickness','Uniformity_of_Cell_Size','Uniformity_of_Cell_Shape','Marginal_Adhesion', 'Single_Epithelial_Cell_Size', 'Bare_Nuclei', 'Bland_Chromatin', 'Normal_Nucleoli', 'Mitoses']
X = df[feature_cols]  
y = df['Malignant_Cancer']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=3)
```

```python
# C. Use “Decision Tree Classifier” to predict Cancer based on the training/testing datasets that you built 
# in part (b). Then, calculate and report the accuracy of your classifier.
dt = DecisionTreeClassifier(random_state=3)
dt.fit(X_train, y_train)
y_predict = dt.predict(X_test)
accuracy = accuracy_score(y_test, y_predict)
print('Decision Tree accuracy:', accuracy)
```

```python
res = []
bootstarp_size = 0.8 * len(X_train)
bootstarp_size = int(bootstarp_size)
for i in range(19):
    # 1. make a bootstarp sample of the original “Training” Dataset (build in part(b)) with the size of
    # bootstarp_size = 0.8*(Size of the original training dataset).
    resample(X_train, n_samples = bootstarp_size , random_state=i , replace = True)
    # 2. Define and train a new base decision tree classifier on this dataset in each iteration:
    Base_DecisionTree = DecisionTreeClassifier(random_state=3)
    Base_DecisionTree.fit(X_train, y_train)
    # 3. Perform prediction using “this base classifier” on the original “Testing” Dataset X_test 
    # (build in part(b)), and save the prediction results for all testing samples.
    prediction = Base_DecisionTree.predict(X_test)
    res.append(prediction)
```

```python
# D. Now, we want to perform a new Ensemble Learning method called “Bagging” based on Voting on 19 decision tree classifiers.
res = []
bootstarp_size = 0.8 * len(X_train)
bootstarp_size = int(bootstarp_size)
for i in range(19):
    # 1. make a bootstarp sample of the original “Training” Dataset (build in part(b)) with the size of
    # bootstarp_size = 0.8*(Size of the original training dataset).
    resample(X_train, n_samples = bootstarp_size , random_state=i , replace = True)
    # 2. Define and train a new base decision tree classifier on this dataset in each iteration:
    Base_DecisionTree = DecisionTreeClassifier(random_state=3)
    Base_DecisionTree.fit(X_train, y_train)
    # 3. Perform prediction using “this base classifier” on the original “Testing” Dataset X_test 
    # (build in part(b)), and save the prediction results for all testing samples.
    prediction = Base_DecisionTree.predict(X_test)
    res.append(prediction)
print(res)
# Checking if I have 19 classifiers
print(len(res))
```

```python
# After finishing the “for” loop, you should have 19 different predictions for EACH sample 
# in your testing set. Then, Perform Voting to make the final decision on each data sample 
# based on the votes of all 19 classifiers.
new_df = pd.DataFrame(res)
voting = []
for column in new_df:
    voting.append(new_df[column].value_counts().idxmax())
# Finally, calculate and report the final accuracy of your Bagging (Voting) method.
accuracy_of_voting = accuracy_score(y_test, voting)
print('Final accuracy of my bagging method:', accuracy_of_voting)
```

```python
# E. Use scikit-learn “Random Forest” classifier to predict Cancer based on the training/testing datasets that 
# you built in part (b). Then, calculate and report the accuracy of your classifier. Use this command to import 
# and define your classifier:
my_RandomForest = RandomForestClassifier(n_estimators = 19, bootstrap = True, random_state=3)
# Similar to previous syntax, use my_RandomForest.fit  for training your random forest classifier 
# and my_RandomForest.predict for prediction.
my_RandomForest.fit(X_train, y_train)
y_predict = my_RandomForest.predict(X_test)
accuracy = accuracy_score(y_test, y_predict)
print('Random Forest Classifier accuracy:', accuracy)
```
